```python
from pykrx import stock
import pandas as pd
```

```python
tickers = ["002810", "003920", "004380", "011280", "016580", "007340", "000370", "001510", "033270", "004710", "001500"]
df = pd.DataFrame()
for ticker in tickers:
    df[ticker] = stock.get_market_ohlcv_by_date("20010101", "20191231", ticker, 'm')['종가']
```

```python
df
```

```python
import copy
ret = copy.deepcopy(df)
for i in range(1, df.shape[0]):
    ret.iloc[i] = df.iloc[i]/df.iloc[i-1]
ret = ret.iloc[1:]
```

```python
ret
```

```python


count_UP = 0
count_DOWN = 0
count_win = 0
upMean = 0
downMean = 0
for ticker in tickers:
    for i in range(6, ret.shape[0]-1):

        if ret[ticker].iloc[i-1]:#*ret[ticker].iloc[i-2]*ret[ticker].iloc[i-3] > 1:
            print('--'*50)
            print(f"과거 3개월 수익률 : {ret[ticker].iloc[i-1]*ret[ticker].iloc[i-2]*ret[ticker].iloc[i-3]}, {ticker}")
            print(f"앞으로 3개월 수익률 {ret[ticker].iloc[i]*ret[ticker].iloc[i+2]*ret[ticker].iloc[i+3]}")
            upMean += ret[ticker].iloc[i]*ret[ticker].iloc[i+2]*ret[ticker].iloc[i+3]
            count_UP += 1
            if ret[ticker].iloc[i]*ret[ticker].iloc[i+2]*ret[ticker].iloc[i+3] > 1:
                count_win += 1
            print('--'*50)
                
```
