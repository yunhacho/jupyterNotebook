```python
import pykrx.stock

class Dataset():
    
    def __init__(self, stock='005930', date=('20190101', '20191231') , theta=10):
        self.df = pykrx.stock.get_market_ohlcv_by_date(date[0], date[1], stock)
        self.observe = theta
        self.theta = theta
    def next(self):
        self.observe += 1
        print(self.observe)

        return self.df.iloc[self.observe - self.theta : self.observe], self.df.iloc[[self.observe]]

ds = Dataset()
```
