```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rawdata = pd.read_csv('./7832디시전트리.csv')
rawdata
```

```python
del rawdata['지원사업명']
```

```python
rawdata.info()
```

```python
rawdata.fillna(0)
```

```python
rawdata['지원금액'] =rawdata['지원금액'] / 1000000
```

```python
rawdata['당년매출액']= rawdata['당년매출액'] / 1000000
```

```python
rawdata['당년투자금액'] = rawdata['당년투자금액'] / 1000000
rawdata['후년매출액'] = rawdata['후년매출액'] / 1000000
rawdata['후년투자금액'] = rawdata['후년투자금액'] / 1000000
```

```python
rawdata
```

```python
rawdata = rawdata.fillna(0)
```

```python
rawdata
```

```python
from sklearn.model_selection import train_test_split as spliter
X_train, X_test, Y_train, Y_test = spliter(rawdata, rawdata[['후년매출액', '후년고용' , '후년직원', '후년투자금액']], test_size = 0.2, random_state=13)

from sklearn import tree
tree_reg = tree.DecisionTreeRegressor()
tree_reg.fit(X_train,Y_train)
```

```python
tree_reg.score(X_train, Y_train)
```

```python
testresult = tree_reg.predict(X_test)
```
