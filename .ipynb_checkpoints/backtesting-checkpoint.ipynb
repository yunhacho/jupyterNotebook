```python
raw_data[ticker["종목코드"].iloc[1]].iloc[1]
```

```python
for i in range(ncol): # # of tickers
    for j in range(nrow):
        if j == 0:
            continue
        prev = raw_data[ticker["종목코드"].iloc[i]].iloc[j-1]
        today = raw_data[ticker["종목코드"].iloc[i]].iloc[j]
        if np.isnan(prev) or np.isnan(today):
            continue
        ret[ticker["종목코드"].iloc[i]].iloc[j]= (raw_data[ticker["종목코드"].iloc[i]].iloc[j] - raw_data[ticker["종목코드"].iloc[i]].iloc[j-1]) / raw_data[ticker["종목코드"].iloc[i]].iloc[j-1]# = (raw_data[ticker["종목코드"].iloc[i]].iloc[i] - raw_data[ticker["종목코드"].iloc[i]].iloc[j-1]) / raw_data[ticker["종목코드"].iloc[i]].iloc[j-1]
```
