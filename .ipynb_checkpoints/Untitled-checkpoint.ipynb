```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('./titanic/train.csv')
```

```python
raw_data.info()
```

```python
raw_data
```

```python
#sex column을 binary digit으로 변경
_ = []
for each in raw_data['Sex']:
    if each == 'male':
        _.append(1)
    elif each == 'female':
        _.append(0)
    else:
        _.append(np.nan)
```

```python
raw_data['Sex'] = _
raw_data['Survived'] = raw_data['Survived'].astype('int')
raw_data['Sex'] = raw_data['Sex'].astype('float')
raw_data['SibSp'] = raw_data['SibSp'].astype('float')
raw_data['Parch'] = raw_data['Parch'].astype('float')
raw_data['Fare'] = raw_data['Fare'].astype('float')

raw_data.info()


```

```python
raw_data = raw_data[raw_data['Age'].notnull()]
raw_data = raw_data[raw_data['SibSp'].notnull()]
raw_data = raw_data[raw_data['Parch'].notnull()]
raw_data = raw_data[raw_data['Fare'].notnull()]

raw_data.info()
```

```python
train_pre = raw_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
train_pre.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(train_pre, raw_data[['Survived']], test_size = 0.2, random_state=13)
```

```python
from sklearn import tree
tree_clf = tree.DecisionTreeClassifier(max_depth=3, random_state=13)
tree_clf.fit(X_train, Y_train)
```

```python
print(tree_clf.score(X_train,Y_train))
```
