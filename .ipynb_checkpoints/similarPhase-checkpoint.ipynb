```python
def getcov(dataset_period):
    vcm_period = np.identity(len(tickers))
    for i in range(len(tickers)):
        for j in range(len(tickers)):
            vcm_period[i][j] = np.cov(dataset_period[tickers[i]].astype(np.float32) , dataset_period[tickers[j]].astype(np.float32))[0,1]
    
    return vcm_period
```

```python
dataset_ret.loc[bsns_date[0]:bsns_month_boundary[0]]
```

```python
vcm_over_periods = []
for i in ran(bsns_month_boundary):
    if i == 0:
        dataset_ret_period = dataset_ret.loc[bsns_date[0]:bsns_month_boundary[i]]
    else:
        dataset_ret_period = dataset_ret.loc[bsns_month_boundary[i-1]:bsns_month_boundary[i]]
    
    #print("******************",i,"*****************")
    #print(getcov(dataset_period))
    #print( pd.DataFrame(getcov(dataset_period)) )
    vcm_over_periods.append(getcov(dataset_ret_period))
    
```

```python
vcm_over_periods[0].shape
```

```python
def distance_vcm(vcm_origin, vcm_target):
    if vcm_origin.shape != vcm_target.shape:
        raise Exception("VCM 크기가 서로 다름!")
        return
    
    loss = 0
    for i in range(vcm_origin.shape[0]):
        for j in range(vcm_origin.shape[1]):
            loss += (vcm_origin[i][j] - vcm_target[i][j])**2
    return loss**(1/2)
    
```

```python
distance_vcm(vcm_over_periods[0], vcm_last1month)
```

```python
losses = []
for i in range(len(vcm_over_periods)):
    print(distance_vcm(vcm_over_periods[i], vcm_last1month))
    losses.append(distance_vcm(vcm_over_periods[i], vcm_last1month))
```
