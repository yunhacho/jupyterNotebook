```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func_rsi(rs):
    rsi = rs/(1+rs)
    return rsi
```

```python
seq = np.arange(0,10,0.001)
n = seq.shape[0]
seq_y = np.zeros(n)
```

```python
for i in range(n):
    seq_y[i] = func_rsi(seq[i])
```

```python
plt.figure("RSI CHART")
plt.plot(seq,seq_y)
plt.xlabel("RS")
plt.ylabel("RSI")
plt.show()
```

```python
def distance(c):
    return (2*(1-c))**(1/2)
```

```python
seq_ = np.arange(-1,1,0.001)
n = seq_.shape[0]
seq_y_ = np.zeros(n)
```

```python
for i in range(n):
    seq_y_[i] = distance(seq_[i])
    
```

```python
plt.figure()
plt.plot( seq_,seq_y_)
plt.show()
```
