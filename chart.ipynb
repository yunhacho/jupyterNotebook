```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func_rsi(rs):
    rsi = rs/(1+rs)
    return rsi
```

```python
seq = np.arange(0,10,0.001)
n = seq.shape[0]
seq_y = np.zeros(n)
```

```python
for i in range(n):
    seq_y[i] = func_rsi(seq[i])
```

```python
plt.figure("RSI CHART")
plt.plot(seq,seq_y)
plt.xlabel("RS")
plt.ylabel("RSI")
plt.show()
```
